```python
for x in data:
    x['Gender'] = x['Gender'].replace({'Male':1,'Female':0})
    x['Foreign_schooled'] = x['Foreign_schooled'].replace({'Yes':1,'No':0})
    x['Previous_IntraDepartmental_Movement'] = x['Previous_IntraDepartmental_Movement'].replace({'Yes':1,'No':0})
    x['Past_Disciplinary_Action'] = x['Past_Disciplinary_Action'].replace({'Yes':1,'No':0})
    x['No_of_previous_employers'] = x['No_of_previous_employers'].replace({'More than 5':6})
```

```python
dict_state = {"SW":['LAGOS', 'ONDO', 'OSUN', 'EKITI', 'OYO', 'OGUN'], 
              "SS":['CROSS RIVER','AKWA IBOM', 'BAYELSA', 'EDO', 'DELTA', 'RIVERS'],
              "SE":['ANAMBRA', 'IMO', 'ABIA', 'EBONYI', 'ENUGU'], 
              "NE":['TARABA', 'ADAMAWA', 'BAUCHI', 'YOBE', 'BORNO', 'GOMBE', 'ADAMAWA'],
              "NW":['KADUNA','KANO', 'ZAMFARA','SOKOTO', 'JIGAWA','KATSINA', 'KEBBI'],
              "NC":['NIGER', 'KOGI', 'BENUE', 'NASSARAWA', 'KWARA', 'PLATEAU', 'FCT']}
```

```python
for x in data:
    state_list = list(x["State_Of_Origin"])

    for i in range(len(state_list)):
        for k, v in enumerate(dict_state.items()):
            if (state_list[i] in list(v)[1]):
                state_list[i] = (list(v)[0])
                
    x["State_Geopolitical_zone"] = state_list
```

```python
for x in data:
    trainings_attended = []
    for i in range(len(x)):  
        if(int(x.loc[i]["Trainings_Attended"]) > 2):
            trainings_attended.append("More than two")
        else:
            trainings_attended.append("Two")
    x['all_trainings_attended'] = trainings_attended
```

```python
for x in data:
    age_range = []
    for i in range(len(x)):  
        if(int(x.loc[i]["age"]) < 18 ):
            age_range.append("0-18")
        elif(int(x.loc[i]["age"]) > 18 and int(x.loc[i]["age"]) < 36):
            age_range.append("19-36")
        elif(int(x.loc[i]["age"]) > 36 and int(x.loc[i]["age"]) < 65):
            age_range.append("37-64")
        else:
            age_range.append("66-above")
    x['age_range'] = age_range
```

```python
train = pd.get_dummies(train, columns = ['State_Geopolitical_zone'])
test = pd.get_dummies(test, columns = ['State_Geopolitical_zone'])

train = pd.get_dummies(train, columns = ['Channel_of_Recruitment'])
test = pd.get_dummies(test, columns = ['Channel_of_Recruitment'])

train = pd.get_dummies(train, columns = ['Marital_Status'])
test = pd.get_dummies(test, columns = ['Marital_Status'])

train = pd.get_dummies(train, columns = ['Division'])
test = pd.get_dummies(test, columns = ['Division'])

train = pd.get_dummies(train, columns = ['all_trainings_attended'])
test = pd.get_dummies(test, columns = ['all_trainings_attended'])

train = pd.get_dummies(train, columns = ['Qualification'])
test = pd.get_dummies(test, columns = ['Qualification'])

train = pd.get_dummies(train, columns = ['age_range'])
test = pd.get_dummies(test, columns = ['age_range'])
```

```python
MM = MinMaxScaler()
train[['Last_performance_score','Training_score_average','year_difference']] = MM.fit_transform(X = train[['Last_performance_score','Training_score_average', 'year_difference']])
test[['Last_performance_score','Training_score_average', 'year_difference']] = MM.fit_transform(X = test[['Last_performance_score','Training_score_average', 'year_difference']]) 
```

```python
y = train['Promoted_or_Not']
X = train.drop(['EmployeeNo','Year_of_birth', 'Year_of_recruitment', "all_trainings_attended_0", "all_trainings_attended_1",'age_range_37-64','Promoted_or_Not', 'State_Of_Origin'], axis=1)

test_1 = test.drop(['EmployeeNo','Year_of_birth', 'Year_of_recruitment', "all_trainings_attended_0", "all_trainings_attended_1",'age_range_37-64','State_Of_Origin'], axis=1)
```

```python
def reg_pred(algo):
    
    # Create training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=50)

    # Create the regressor: reg_all
    reg_all = algo

    # Fit the regressor to the training data
    reg_all.fit(X_train, y_train)

    # Predict on the test data: y_pred
    y_pred = reg_all.predict(X_test)

    # Compute and print R^2 and RMSE
    print("R^2: {}".format(reg_all.score(X_test, y_test)))
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    print("Root Mean Squared Error: {}".format(rmse))
```

```python
init_adab = AdaBoostClassifier(learning_rate = 0.4)
reg_pred(init_adab)

init_adab.fit(X,y)
result = init_adab.predict(test_1)

submission = pd.DataFrame({"EmployeeNo": test["EmployeeNo"], "Promoted_or_Not": result})
submission.to_csv("result_submission_20.csv", index=False)
```
